## Data Analysis with Visualization


Creating visualizations to explore and understand the data further.

### Scatterplot

```python
plt.scatter(x=df['Time on App'], y=df['Time on Website'])
plt.xticks(label=50)
plt.xlabel('Time on App')
plt.ylabel('Time on Website')
plt.title('Time Investment')
plt.show()
```

```python
plt.scatter(x=df['Length of Membership'], y=df['Yearly Amount Spent'])
plt.xticks(label=50)
plt.xlabel('Length of Membership')
plt.ylabel('Yearly Amount Spent')
plt.title('Expenditure')
plt.show()
```

### Boxplot 

```python
sns.boxplot(x=df['Yearly Amount Spent'])
plt.xlabel('Length of Membership')
plt.ylabel('Yearly Amount Spent')
plt.title('Expenditure')
plt.show()
```

### Barplot

```python
sns.barplot(x='Time on Website', y='Yearly Amount Spent',data=df) 
```

### Joint Plot

```python
sns.jointplot(x=df['Time on Website' ],y=df['Yearly Amount Spent'])
```

```python
sns.jointplot(x=df['Time on App'],y=df['Yearly Amount Spent'])
```

```python
sns.jointplot(x=df['Time on App'],y=df['Yearly Amount Spent'],kind='hex')
```

### Pair Plot 

```python
sns.pairplot(df, kind= 'reg')
```

### Violine Plot

```python
sns.violinplot(data=df, x='Time on App',split=True,inner="stick")
```

### Area plot 

```python
df.plot.area()
```
